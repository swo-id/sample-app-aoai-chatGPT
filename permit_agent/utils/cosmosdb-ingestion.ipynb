```python
import uuid

def transform_permit_data(item_data: dict) -> dict:
    """
    Transform raw permit data into the format required for Cosmos DB.
    
    Args:
        item_data: List of permit items with file_path, permit_type, issue_date, 
                   expiration_date, and summary fields.
    
    Returns:
        List of transformed permit data dictionaries.
    """
    
    data = {
            "id": uuid.uuid4().hex,
            "filepath": item_data["file_path"],
            "documentTitle": item_data['file_path'].split('/')[-1],
            "organization": item_data["file_path"].split('/')[2].split('_')[0],
            "keywords": item_data["file_path"].split('/')[3:],
            "permitType": item_data["permit_type"],
            "permits": [
                {
                    "issueDate": item_data['issue_date'],
                    "expirationDate": item_data['expiration_date'],
                    "permitSummary": item_data['summary'],
                    "permitNumber": item_data['permit_number'].replace(' ', '')
                }
            ]
        }
    
    return data

def transform_permit_data_plo(item_data: dict) -> dict:
    """
    Transform raw permit data into the format required for Cosmos DB.
    
    Args:
        item_data: List of permit items with file_path, permit_type, issue_date, 
                   expiration_date, and summary fields.
    
    Returns:
        List of transformed permit data dictionaries.
    """
    
    data = {
            "id": uuid.uuid4().hex,
            "filepath": item_data["file_path"],
            "documentTitle": item_data['file_path'].split('/')[-1],
            "organization": item_data["file_path"].split('/')[2].split('_')[0],
            "keywords": item_data["file_path"].split('/')[3:],
            "permitType": item_data["permit_type"],
            "permits": [
                {
                    "issueDate": item_data['issue_date'],
                    "expirationDate": item_data['expiration_date'],
                    "permitSummary": item_data['summary'],
                    "permitNumber": item_data['permit_number'].replace(' ', ''),
                    "installation" : item_data['installation'],
                }
            ]
        }
    
    return data
```
