## Data Export Manual Batching

```python
import time

subset_run_items = dataset_run_items[:30]

prompt_inputs = []
outputs = []
latencys = []
scores = []

BATCH_SIZE = 5

for i in range(0, len(subset_run_items), BATCH_SIZE):
    batch = subset_run_items[i:i+BATCH_SIZE]
    print(f"Processing batch {i//BATCH_SIZE + 1} with {len(batch)} items...")
    for item in batch:
        trace = langfuse.api.trace.get(item.trace_id).dict()
        prompt_inputs.append(trace["input"])
        outputs.append(trace["output"])
        latencys.append(trace["latency"])
        scores.append(trace["scores"][0]["value"])
    time.sleep(5)  # Wait 5 seconds between batches
```

```python
import pandas as pd

data = {
    "prompt_input": prompt_inputs,
    "output": outputs,
    "latency": latencys,
    "score": scores
}

df = pd.DataFrame(data)
df.to_csv("data/trace_data_batch_1_30.csv", index=False)
```
